```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Generate Synthetic Customer Data
np.random.seed(42)  # for reproducibility
n_customers = 1000

customer_ids = np.arange(1, n_customers + 1)
tenure = np.random.randint(1, 72, n_customers)  # Months of service
monthly_charges = np.random.uniform(20, 120, n_customers)
total_charges = tenure * monthly_charges + np.random.normal(0, 50, n_customers) # Adding some noise
contract_type = np.random.choice(['Month-to-month', 'One year', 'Two year'], n_customers)
online_security = np.random.choice(['Yes', 'No'], n_customers)
tech_support = np.random.choice(['Yes', 'No'], n_customers)
internet_service = np.random.choice(['DSL', 'Fiber optic', 'No'], n_customers)

# Simulate churn based on some conditions (these are the hidden patterns we want to uncover)
churn = np.zeros(n_customers, dtype=int)
for i in range(n_customers):
    if contract_type[i] == 'Month-to-month' and tenure[i] < 24 and monthly_charges[i] > 70:
        churn[i] = np.random.choice([0, 1], p=[0.3, 0.7]) # Higher chance of churn
    elif internet_service[i] == 'Fiber optic' and tech_support[i] == 'No':
        churn[i] = np.random.choice([0, 1], p=[0.4, 0.6])
    elif tenure[i] > 60 and contract_type[i] == 'Two year':
        churn[i] = np.random.choice([0, 1], p=[0.8, 0.2]) # Lower chance of churn
    else:
        churn[i] = np.random.choice([0, 1], p=[0.85, 0.15]) # Base churn rate

# Create a Pandas DataFrame
data = pd.DataFrame({
    'CustomerID': customer_ids,
    'Tenure': tenure,
    'MonthlyCharges': monthly_charges,
    'TotalCharges': total_charges,
    'ContractType': contract_type,
    'OnlineSecurity': online_security,
    'TechSupport': tech_support,
    'InternetService': internet_service,
    'Churn': churn
})

# 2. Data Preprocessing
# Handle categorical features using one-hot encoding
data = pd.get_dummies(data, columns=['ContractType', 'OnlineSecurity', 'TechSupport', 'InternetService'], drop_first=True)

# Drop CustomerID as it's not a predictive feature
data = data.drop('CustomerID', axis=1)

# Separate features (X) and target (y)
X = data.drop('Churn', axis=1)
y = data['Churn']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Model Training
# Initialize a Random Forest Classifier (you can experiment with other models)
model = RandomForestClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)

# 4. Model Evaluation
# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(
    y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Churn', 'Churn'], yticklabels=['No Churn', 'Churn'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# 5. Feature Importance (to get insights into potential patterns)
feature_importances = model.feature_importances_
feature_names = X.columns

sorted_indices = np.argsort(feature_importances)[::-1]

plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.bar(range(X.shape[1]), feature_importances[sorted_indices], align="center")
plt.xticks(range(X.shape[1]), feature_names[sorted_indices], rotation=90)
plt.tight_layout()
plt.show()
```
